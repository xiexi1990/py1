```python
import numpy as np
import random
from scipy import sparse

random.seed(1234)

row = 40
col = 40
A = np.zeros((row,col))
n = 1001
for i in range(0,row):
    rnz = random.randint(0,5)
    if i == 18:
        rnz = 23
    location = random.sample(list(range(0,col)), rnz)
    for j in sorted(location):
        A[i, j] = n
        n += 1

S = sparse.csr_matrix(A)
sparse.save_npz("npz1.npz", S)
print(S.nnz)

for i in range(row):
    for j in range(col):
        print(int(A[i, j]), end=" ")
    print("")                  
```
